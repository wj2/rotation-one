## Creation of a rate model to explain specific results in ITC recordings ##

Two main parts: 
1. Fitting of a transfer function to the assumed F-I curve
  * we assume that the input currents to the neurons are gaussian
  * we will try several different base transfer functions, beginning with a sigmoid
2. Extraction of reasonable original synaptic weights
  * we hope to obtain a gaussian distribution of weights for each neuron

Now, we'll decide how to initialize our synaptic weights. 

We know the resting activity of the network. We also know the activity of the network in response to a feedforward input of unknown magnitude. 

```python
print stim_allnov.shape

# mean activity with no stimulus
noise_m, noise_std = noise_allnov.mean(), noise_allnov.std()
# mean activity with stimulus
stim = 21
stim_m, stim_std = stim_allnov[:, stim].mean(), stim_allnov[:, stim].std()
astim_m, astim_std = stim_allnov.mean(), stim_allnov.std()

print 'noise mean: {} \nnoise std: {} \nstim mean: {} \nstim std: {}'.format(noise_m, noise_std,
                                                                             stim_m, stim_std)
print astim_m, astim_std
```

We are using $\dot{\vec{v}} = -\vec{v} + \phi(u + \mathbf{w}\vec{v})$ where $\phi$ is our nonlinearity (the tanh function from above in this case), $u$ is a scalar representing feedforward input, $\mathbf{w}$ is an $m \times m$ weight matrix, and $\vec{v}$ is an $m \times 1$ vector of recurrent activity.

We are going to make the assumption that $u$ has a gaussian distribution with constant variance for a given stimulus. From the above, we'll make the (dubious) assumption that $\vec{v}$ has an initial condition given by a gaussian with $\mu = 6.7$ and $\sigma = 8.676$. 

We want to get a distribution of weights for the initial condition as well. We solve $y = \phi(wy)$ and obtain $w = \frac{\phi^{-1}(y)}{y}$.

```python
a, b, c, d = tvars
print a, b, c, d
wnoise_mean = (((np.arctanh((noise_m - b)/a) - c) / d) - all_currs.min()) / noise_m
wstim_mean = (((np.arctanh((stim_m - b)/a) - c) / d) - all_currs.min()) / stim_m
w2noise_mean = (((np.arctanh((noise_m - b)/a) - c) / d)) / noise_m
w2stim_mean = (((np.arctanh((stim_m - b)/a) - c) / d)) / stim_m

print 'noise mean starting weight: {}'.format(wnoise_mean)
print 'stim mean starting weight: {}'.format(wstim_mean)
print 'noise mean starting weight w/o adjust: {}'.format(w2noise_mean)
print 'stim mean starting weight w/o adjust: {}'.format(w2stim_mean)
```

We will use $1 / m$ for the standard deviation of our initial recurrent weights. 

Now we must decide the static weight on our feedforward input such that when we recieve a gaussian feedforward input of some mean $\mu_i$, we increase our mean activity to something reasonably within the range of what we've observed for post-stimulus activity levels -- we'll choose 12, tentatively. 

```python
r_activity = 12

estin = ((np.arctanh((r_activity - b)/a) - c) / d) - all_currs.min() - wnoise_mean*noise_m
print estin
```

This is our estimated input $u$ for a reasonable stimulus. 

We will center a gaussian on $u$ with some small standard deviation. 

We will now fully specify our model:

$$ \tau_r\dot{v} = -v + \phi(u + \mathbf{w}v) $$
$$ \tau_w\dot{\mathbf{w}} = (v - \alpha\bar{v})(v - \bar{v})^T $$
$$ \tau_a\dot{\bar{v}} = -\bar{v} + v $$

where $v$ is an $m \times 1$ vector representing the firing rates of the $m$ units, $\bar{v}$ is an $m \times 1$ is a vector representing a windowed average of each unit's firing rate, $\phi$ is the transfer function as specified above, $u$ is a scalar giving the feedforward input to the network, $\mathbf{w}$ is an $m \times m$ matrix giving the recurrent weights of the network, and $\alpha$ is a scalar.

Next, we will try this simulation using the BCM rule, formulated as follows:

$$ \tau_r\dot{v} = -v + \phi(u + \mathbf{w}v) $$
$$ \tau_w\dot{\mathbf{w}} = vv^T * (v - \theta) $$
$$ \tau_a\dot{e} = -e + v $$
$$ \tau_a\dot{s} = -s + v^2 $$

$$ \theta = e + \lambda\sqrt{s - e^2} $$

```python
reload(rmn)

# initial conditions:
m = 10.
n = 1.
v_std = .4
# v0 = get_gaussian(m, 1, noise_m-1, v_std)
v0 = np.random.exponential(scale=noise_m, size=m).reshape(m, 1)
# v0 = np.ones((m, 1)) * noise_m
# w0 = get_gaussian(m, m, wnoise_mean / m, 1/(2*m))
w0 = np.ones((m, m)) * (wnoise_mean*.937973) / m
# w0 = np.random.exponential(scale=wnoise_mean/m, size=m*m).reshape(10, 10)
inweight = np.ones((m, 1))
tau_r = 10 # ms
tau_w = 20000 # ms
tau_a = 50 # ms
lam = 1.16
weight_sat = 1
v_bar = v0

const_in = lambda x: 0
u_func = iofuncs.pattern_gaussians([estin], [.1], 1000, 300, 0, 0)

net = rmn.RateNetworkBCM(m, n, tanh_fset, tau_r, tau_a, tau_w, weight_sat, lam)
ts, ys = net.simulate_euler(v0, inweight, w0, v_bar, v_bar**2, 1000, .01, u_func)
net.plot_simulation(ys[:-1, :], tcourse=ts[:-1], figsize=(30, 20))
```
